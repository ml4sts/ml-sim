# Demographic Sampler

```python
testPop = mlsim.Population()
df = testPop.sample(5000)
probA = sum(df['a'])/5000
probZ = sum(df['z'])/5000

assert abs(probA - testPop.demographic_sampler.get_rho_a()) < .02
assert abs(probZ - testPop.demographic_sampler.get_rho_z()[0]) < .02
```

```python
sampleSizes = [100,250,500,600,700,800,900,1000,2000,3000,4000,5000]
df = pd.DataFrame(columns = sampleSizes)
df2 = pd.DataFrame(columns = sampleSizes)
testPop = mlsim.Population()
for i in sampleSizes:
    Errors_A = []
    Errors_Z = []
    for c in range(10):
        temp_df = testPop.sample(i)
        probA = sum(temp_df['a'])/i
        probZ = sum(temp_df['z'])/i
        Errors_A.append(abs(probA - testPop.demographic_sampler.get_rho_a()))
        Errors_Z.append(abs(probZ - testPop.demographic_sampler.get_rho_z()[0]))
    df[i] = Errors_A
    df2[i] = Errors_Z
sns.catplot(palette="YlGnBu_d", height=6, aspect=1, kind="box", data=df)
sns.catplot(palette="YlGnBu_d", height=6, aspect=1, kind="box", data=df2)
```
